 ## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
```

- **Pandas**: Data loading and manipulation
- **Matplotlib**: Core plotting library
- **Seaborn**: Statistical visualization with enhanced styling
- **Warnings**: Suppressed to keep output clean

## Load Dataset

```python
df = sns.load_dataset('iris')
```

- **Source**: Loaded via seaborn.load_dataset()
- **Features**: Sepal/petal length & width measurements in cm
- **Samples**: 150 iris flower observations
- **Classes**: 3 species (setosa, versicolor, virginica)

## Exploratory Data Analysis
Understanding data structure, distributions, and relationships

## Basic Inspection
- Dataset dimensions and column names
- First few rows to understand data format
- Data types and missing value counts
- Summary statistics (mean, std, min, max, quartiles)

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for idx, col in enumerate(numeric_cols):
    ax = axes[idx // 2, idx % 2]
    sns.boxplot(data=df, x='species', y=col, ax=ax, palette='Set2')
    ax.set_title(f'Box Plot of {col.replace("_", " ").title()}', fontsize=11, fontweight='bold')
    ax.set_xlabel('Species')
    ax.set_ylabel(f'{col.replace("_", " ").title()} (cm)')

plt.tight_layout()
plt.show()
```

## Visualization - Box Plots (Outliers)
- Quartiles and median for each feature
- Whiskers showing data range
- Outliers displayed as individual circles
- Grouped by iris species

```python
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, col in enumerate(numeric_cols):
    ax = axes[idx // 2, idx % 2]
    sns.histplot(data=df, x=col, hue='species', kde=True, ax=ax, bins=20)
    ax.set_title(f'Distribution of {col.replace("_", " ").title()}', fontsize=11, fontweight='bold')
    ax.set_xlabel(f'{col.replace("_", " ").title()} (cm)')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Visualization - Distributions (Histograms)
- Frequency distribution for each feature
- KDE curves showing probability density
- Species overlaid for comparison
- Reveals distinct distribution patterns by species

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', s=100)
plt.title('Sepal Length vs Petal Length', fontsize=12, fontweight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='sepal_width', y='petal_width', hue='species', s=100)
plt.title('Sepal Width vs Petal Width', fontsize=12, fontweight='bold')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```

## Visualization - Scatter Plots
- Feature relationships and correlation patterns
- Species separation visible through color coding
- Pairwise comparisons: sepal vs petal dimensions
- Helps identify linear relationships and clustering
